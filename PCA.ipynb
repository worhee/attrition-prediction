```python
import csv
import random
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
import matplotlib.pyplot as plt
from random import shuffle, seed
```

```python
with open('PREPROCESSED_WA_Fn-UseC_-HR-Employee-Attrition.csv', 'r') as f:

    lines = f.readlines()
    y_at1, X_at1 = [], []
    for line in lines[1:238]:
        line = line[:-1]
        splited = line.split(',')
        y_at1.append([int(splited[0])])
        X_at1.append([int(value) for value in splited[1:]])
    y_at0, X_at0 = [], []
    for line in lines[238:]:
        line = line[:-1]
        splited = line.split(',')
        y_at0.append([int(splited[0])])
        X_at0.append([int(value) for value in splited[1:]])


random.Random(0).shuffle(X_at1)
random.Random(0).shuffle(X_at0)

X_train = np.array(X_at1[0:150] + X_at0[0:150])
y_train = np.array(y_at1[0:150] + y_at0[0:150])
X_test = np.array(X_at1[150:] + X_at0[150:])
y_test = np.array(y_at1[150:] + y_at0[150:])
```

```python
pca = PCA(n_components=2)
pca.fit(X_train)
var = pca.explained_variance_ratio_
var_cum=np.cumsum(np.round(var, decimals=4)*100)
var
#var_cum
```
